## 2. Build the ranked queue (writes the CSV)

*Code the score, rank everything, write work/outputs/baseline_action_score.csv.*

```python
import pandas as pd

# Load data
df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")

# CTR
df["ctr"] = df["clicks_90d"] / df["impressions_90d"].replace(0, 1)

# Score
df["baseline_score"] = (
    df["impressions_90d"].fillna(0) * 0.4 +
    (50 - df["avg_position"].fillna(50)) * 2 +
    df["engagement_rate"].fillna(0) * 5 +
    df["trend_pct"].fillna(0)
)

# Reason code
df["reason_code"] = "normal"

df.loc[df["avg_position"] > 20, "reason_code"] = "position_slipping"
df.loc[df["trend_pct"] < -20, "reason_code"] = "traffic_dropping"
df.loc[df["engagement_rate"] < 1, "reason_code"] = "low_engagement"

# Rank
df = df.sort_values("baseline_score", ascending=False)

# Save CSV
df.to_csv("baseline_action_score.csv", index=False)


print("Done!")
print(df[["content_id","baseline_score","reason_code"]].head(20))
```
